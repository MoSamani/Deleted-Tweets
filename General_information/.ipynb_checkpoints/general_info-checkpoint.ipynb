```python
#Dataframe TweetsCov19 
df_tweetscov19 = pd.read_csv(TweetsCov19_Pfad, sep='\t', error_bad_lines=False, names=TweetsCov19_labels)
print('Shape:', df_tweetscov19.shape)
df_tweetscov19.head()
```

```python
#Dataframe TweetsKB 
df_tweetskb = pd.read_csv(TweetsKB_Pfad, sep='\t', error_bad_lines=False, names=TweetsKB_labels)
print('Form :', df_tweetskb.shape)
df_tweetskb.head()
```

```python
df_del_cov19 = df_tweetscov19[df_tweetscov19['Is_deleted'] == True]
df_undel_cov19 = df_tweetscov19[df_tweetscov19['Is_deleted'] == False]
```

# Number of posted and deleted tweets/users in both datasets

#### TweetsCOV19


```python
print('Number of posted tweets: ', df_tweetscov19.shape[0])
print('Number of deleted tweets: ', df_del_cov19.shape[0])
print('Number of non-deleted Tweets: ', df_undel_cov19.shape[0])
print('-----------------------------------')
print('Number of users in posted tweets: ', len(df_tweetscov19.Username.unique()))
print('Number of users in deleted tweets: ', len(df_del_cov19.Username.unique()))
print('Number of users in non-deleted tweets: ', len(df_undel_cov19.Username.unique()))
```

```python
df_del_user =df_del_cov19[['Username']]
df_undel_user =df_undel_cov19[['Username']]

df_cov19_user_undel = df_undel_user[df_undel_user.Username.isin(df_del_user.Username)]
print('Number of users in deleted and none-deleted tweets: ', len(df_cov19_user_undel.Username.unique()))
df_cov19_user_undel = df_undel_user[~df_undel_user.Username.isin(df_del_user.Username)]
print('Number of users in only none-deleted tweets: ', len(df_cov19_user_undel.Username.unique()))
```

```python
df_tweetscov19['deleted'] = df_tweetscov19.apply(lambda row: 1 if (row['Is_deleted']==True) else 0, axis=1)
df_tweetscov19['posted'] = 1
```

## Sentiment

```python
def get_neg_sentiment(txt):
    return float(txt.split()[1])

def get_pos_sentiment(txt):
    return float(txt.split()[0])
```

```python
df_tweetscov19["negative"] = df_tweetscov19.apply(lambda row: get_neg_sentiment(row["Sentiment"]), axis=1)
df_tweetscov19["positive"] = df_tweetscov19.apply(lambda row: get_pos_sentiment(row["Sentiment"]), axis=1)
df_tweetscov19["average"] = df_tweetscov19.apply(lambda row: (row["negative"] + row["positive"]) /2, axis=1)
```
