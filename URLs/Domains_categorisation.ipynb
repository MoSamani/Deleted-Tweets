```python

def get_domain_categorie( domains):
    dom = ''
    categorie = ''
    list_dom = domains.split()
    for i in range(0, len(list_dom)):
        
        categorie=''
        if list_dom[i] != 'nan':
            line = df_tweetscov19_domains_shalla.loc[df_tweetscov19_domains_shalla['domain'] == list_dom[i]]
            line = line.values.tolist()

            try :
                categorie = line[0][1]
            except IndexError:
            
                print('Index out of range!')
                print(list_dom[i])
                print(i)
                print(line)

            
        if isNaN(categorie):
            categorie = str(categorie)
            

        if i == 0 :
            dom = dom + categorie
        else:
            dom = dom + ' ' +categorie

    return dom
```

```python
df_tweetscov19['Domains'] = df_tweetscov19['Domains'].astype(str)

df_tweetscov19['domain_categorie'] = df_tweetscov19.apply(lambda row: get_domain_categorie(row['Domains']) if (row['Domains'] != '0') else '0', axis=1)
print('Domains!!!')
```

```python
#Labels
TweetsCov19_all_domain_categorie_labels = ['Tweet_Id', 'Is_deleted', 'Number of URLS', 'Domains', 'domain_categorie']
#Path
TweetsCov19_all_domain_categorie_path =  '../data/tweetsCov19_all_url_categorized.tsv'

#Dataframe TweetsCov19 
df_tweetscov19 = pd.read_csv(TweetsCov19_all_domain_categorie_path, sep='\t', error_bad_lines=False, names=TweetsCov19_all_domain_categorie_labels)
print('Form', df_tweetscov19.shape)
df_tweetscov19.head()
```

```python
def get_most_used_categories(df): 
    dict_domains_cat = {}
    x = 0
    list_cat = []
    for idx, row in df.iterrows():
        list_cat = (row.values.tolist()[4])
        #print(list_cat)
        list_cat = list_cat.split()
        for i in range(0, len(list_cat)):
            cat = list_cat[i]
            x += 1
            if cat not in dict_domains_cat:
                dict_domains_cat[cat] = 1
            else :
                dict_domains_cat[cat] = dict_domains_cat[cat] +1
        #if x == 50:
        #    break
    return  dict_domains_cat
```

```python
df_tweetscov19['domain_categorie'] = df_tweetscov19['domain_categorie'].astype(str)
TweetsCOV19_dict_domains_categorie = get_most_used_categories(df_tweetscov19)
```

```python
df_tweetscov19_dom_cat = pd.DataFrame(TweetsCOV19_dict_domains_categorie.items(), columns=['Domain_categorie','Number'])
```
