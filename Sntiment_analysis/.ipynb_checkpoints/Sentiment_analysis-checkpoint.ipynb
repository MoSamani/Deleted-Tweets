## TweetsCov19 
### Sentiment analysis (Sentistrengh )

***

```python
def get_positive_seniment(text):
    sentiment_pols = text.split()
    return int(sentiment_pols[0])
```

```python
def get_negative_seniment(text):
    sentiment_pols = text.split()
    return int(sentiment_pols[1])
```

```python
df_tweetscov19['Sentiment'] = df_tweetscov19['Sentiment'].astype(str)
df_tweetscov19['Sentistrengh_pos'] = df_tweetscov19.apply(lambda row: get_positive_seniment(row['Sentiment']) , axis=1)
```

```python
df_tweetscov19['Sentistrengh_neg'] = df_tweetscov19.apply(lambda row: get_negative_seniment(row['Sentiment']) , axis=1)
```

```python
df_tweetscov19['Sentistrengh_average'] = df_tweetscov19.apply(lambda row: (row['Sentistrengh_neg'] + row['Sentistrengh_pos']) , axis=1)
```

```python
df_tweetscov19_sentistrength = df_tweetscov19[['Tweet_Id', 'Timestamp', 'Is_deleted','Sentistrengh_pos', 'Sentistrengh_neg', 'Sentistrengh_average']]
df_tweetscov19_sentistrength.head()
```

```python
pfad_tsv3 = '../data/tweetscov19_sentiment_analysis_sentistrength.tsv'

with open(pfad_tsv3, 'w', newline='', encoding="utf-8") as out_file:
    tsv_writer = csv.writer(out_file, delimiter="\t")
    for idx, row in df_tweetscov19_sentistrength.iterrows():
        tsv_writer.writerow(row.values.tolist())
```
