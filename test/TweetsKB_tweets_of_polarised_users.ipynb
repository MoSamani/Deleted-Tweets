```python
path = '../data/tweetskb_political_polarized_users.tsv'

df_tweetskb_political_polarized_users = pd.read_csv(path, sep='\t', error_bad_lines=False , names=['Username', 'Number_of_liberals', 'Number_of_conservatives', 'Sum_polarised_domains', 'user_political_polarisation'])

print('Shape :', df_tweetskb_political_polarized_users.shape)
df_tweetskb_political_polarized_users.head()
```

```python
path = '../data/tweetskb_science_polarized_users.tsv'

df_tweetskb_science_polarized_users = pd.read_csv(path, sep='\t' , error_bad_lines=False, names=['Username', 'Number_of_antiscience_domains', 'Number_of_proscience_domains', 'Sum_polarised_domains', 'user_science_polarisation'])

print('Shape :', df_tweetskb_science_polarized_users.shape)
df_tweetskb_science_polarized_users.head()
```

```python
path = '../data/tweetskb_moderacy_polarized_users.tsv'

df_tweetskb_moderacy_polarized_users = pd.read_csv(path, sep='\t' , error_bad_lines=False, names=['Username', 'Number_of_moderates', 'Number_of_hardlines', 'Sum_polarised_domains', 'user_moderacy_polarisation'])

print('Shape :', df_tweetskb_moderacy_polarized_users.shape)
df_tweetskb_moderacy_polarized_users.head()
```

```python

df_tweetskb_political_polarized_users = df_tweetskb_political_polarized_users[['Username', 'user_political_polarisation']]
df_tweetskb_science_polarized_users = df_tweetskb_science_polarized_users[['Username', 'user_science_polarisation']]

df_tweetskb_polarized_users = pd.merge(df_tweetskb_political_polarized_users, df_tweetskb_science_polarized_users, on='Username', how='outer')
print('Shape :', df_tweetskb_polarized_users.shape)
df_tweetskb_polarized_users.head()
```

```python
df_tweetskb_moderacy_polarized_users = df_tweetskb_moderacy_polarized_users[['Username', 'user_moderacy_polarisation']]
df_tweetskb_polarized_users = pd.merge(df_tweetskb_polarized_users, df_tweetskb_moderacy_polarized_users, on='Username', how='outer')

print('Shape :', df_tweetskb_polarized_users.shape)
df_tweetskb_polarized_users.head()
```

## Merge tweets polarisation

```python
path = '../data/tweetskb_political_polarized_tweets.tsv'

df_tweetskb_political_polarized_tweets = pd.read_csv(path, sep='\t', error_bad_lines=False, dtype={'Tweet_Id':'object'}, names=['Tweet_Id', 'Number_of_liberals', 'Number_of_conservatives', 'Sum_polarized_domains', 'tweet_political_polarisation'])

print('Shape :', df_tweetskb_political_polarized_tweets.shape)
df_tweetskb_political_polarized_tweets.head()
```

```python
path = '../data/tweetskb_science_polarized_tweets.tsv'

df_tweetskb_science_polarized_tweets = pd.read_csv(path, sep='\t', error_bad_lines=False, dtype={'Tweet_Id':'object'} ,names=['Tweet_Id', 'number_of_antisciens_domains', 'number_of_prosciens_domains', 'Sum_polarized_domains', 'tweet_science_polarisation'])

print('Shape :', df_tweetskb_science_polarized_tweets.shape)
df_tweetskb_science_polarized_tweets.head()
```

```python
path = '../data/tweetskb_moderacy_polarized_tweets.tsv'

df_tweetskb_moderacy_polarized_tweets = pd.read_csv(path, sep='\t', error_bad_lines=False, dtype={'Tweet_Id':'object'}, names=['Tweet_Id', 'Number_of_moderate_domains', 'Number_of_hardline_domains', 'Sum_polarized_domains', 'tweet_moderacy_polarisation'])

print('Shape :', df_tweetskb_moderacy_polarized_tweets.shape)
df_tweetskb_moderacy_polarized_tweets.head()
```

```python
df_tweetskb_political_polarized_tweets = df_tweetskb_political_polarized_tweets[['Tweet_Id', 'tweet_political_polarisation']]
df_tweetskb_science_polarized_tweets = df_tweetskb_science_polarized_tweets[['Tweet_Id', 'tweet_science_polarisation']]
df_tweetskb_moderacy_polarized_tweets = df_tweetskb_moderacy_polarized_tweets[['Tweet_Id', 'tweet_moderacy_polarisation']]

df_tweetskb_political_polarized_tweets.Tweet_Id = df_tweetskb_political_polarized_tweets.Tweet_Id.astype(str) 
df_tweetskb_science_polarized_tweets.Tweet_Id = df_tweetskb_science_polarized_tweets.Tweet_Id.astype(str) 
df_tweetskb_moderacy_polarized_tweets.Tweet_Id = df_tweetskb_moderacy_polarized_tweets.Tweet_Id.astype(str) 

df_tweetskb_polarized_tweets = pd.merge(df_tweetskb_science_polarized_tweets, df_tweetskb_political_polarized_tweets, on= 'Tweet_Id', how='outer' )
df_tweetskb_polarized_tweets = pd.merge(df_tweetskb_polarized_tweets, df_tweetskb_moderacy_polarized_tweets, on= 'Tweet_Id', how='outer' )
```
