```python
#CSV Bibliothek importieren 
import csv
#Pandas importieren und als pd bezeichnen
import pandas as pd
#Numpy importieren und als np bezeichnen 
import numpy as np
#Json Biliothek importieren
import json
#Obere Schranke für die Größe der Daten auf 100,000,000 setzen  
csv.field_size_limit(100000000)
#Maximale Anzahl von Zeilen und Spalten Zeigen 
pd.options.display.max_columns = None
pd.options.display.max_rows = None

pd.options.display.float_format = '{:.2f}'.format
```

# TweetsKB

```python
path = '../data/tweetskb_political_polarized_users_with_atleast_one_polarized_domain.tsv'

df_tweetskb_political_polarized_users = pd.read_csv(path, sep='\t', error_bad_lines=False, names=['Username', 'Number_of_liberals', 'Number_of_conservatives', 'Sum_polarised_domains', 'political_polarisation'])

print('Shape :', df_tweetskb_political_polarized_users.shape)
df_tweetskb_political_polarized_users.head()
```

```python
path = '../data/tweetskb_science_polarized_users_with_atleast_one_polarized_domain.tsv'

df_tweetskb_science_polarized_users = pd.read_csv(path, sep='\t', error_bad_lines=False, names=['Username', 'Number_of_antiscience_domains', 'Number_of_proscience_domains', 'Sum_polarised_domains', 'science_polarisation'])

print('Shape :', df_tweetskb_science_polarized_users.shape)
df_tweetskb_science_polarized_users.head()
```

```python
path = '../data/tweetskb_moderacy_polarized_users_with_atleast_one_polarized_domain.tsv'

df_tweetskb_moderacy_polarized_users = pd.read_csv(path, sep='\t', error_bad_lines=False, names=['Username', 'Number_of_moderates', 'Number_of_hardlines', 'Sum_polarised_domains', 'moderacy_polarisation'])

print('Shape :', df_tweetskb_moderacy_polarized_users.shape)
df_tweetskb_moderacy_polarized_users.head()
```

```python

df_tweetskb_political_polarized_users = df_tweetskb_political_polarized_users[['Username', 'political_polarisation']]
df_tweetskb_science_polarized_users = df_tweetskb_science_polarized_users[['Username', 'science_polarisation']]

df_tweetskb_polarized_users = pd.merge(df_tweetskb_political_polarized_users, df_tweetskb_science_polarized_users, on='Username', how='outer')
print('Shape :', df_tweetskb_polarized_users.shape)
df_tweetskb_polarized_users.head()
```

```python
df_tweetskb_moderacy_polarized_users = df_tweetskb_moderacy_polarized_users[['Username', 'moderacy_polarisation']]
df_tweetskb_polarized_users = pd.merge(df_tweetskb_polarized_users, df_tweetskb_moderacy_polarized_users, on='Username', how='outer')

print('Shape :', df_tweetskb_polarized_users.shape)
df_tweetskb_polarized_users.head()
```

```python
path = '../data/TweetsKB_alltweets_reputation_deletionratio.tsv'

df_tweetskb_alltweets_reputation_deletionratio = pd.read_csv(path, sep='\t', error_bad_lines=False, names=['Tweet_Id', 'Username', 'Followers', 'Friends', 'is_deleted', 'num_posted', 'num_del', 'Deletionratio',  'Reputation'])

print('Form :', df_tweetskb_alltweets_reputation_deletionratio.shape)
df_tweetskb_alltweets_reputation_deletionratio.head()
```

```python
df_tweetskb_alltweets_reputation_deletionratio = df_tweetskb_alltweets_reputation_deletionratio[['Username', 'Deletionratio', 'Reputation', 'is_deleted']]

df_tweetskb_polarized_users = pd.merge(df_tweetskb_polarized_users, df_tweetskb_alltweets_reputation_deletionratio, on='Username', how='inner')
print('Form :', df_tweetskb_polarized_users.shape)
df_tweetskb_polarized_users.head()
```

```python
df_tweetskb_polarized_users = df_tweetskb_polarized_users.replace(np.nan, 0)
df_tweetskb_polarized_users[df_tweetskb_polarized_users['moderacy_polarisation'] == 0].head()
```

```python
pfad_tsv3 = '../data/tweetsKB_polarized_and_neutral_users_deletionratio_reputation.tsv'

with open(pfad_tsv3, 'w', newline='', encoding="utf-8") as out_file:
    tsv_writer = csv.writer(out_file, delimiter="\t")
    for idx, row in df_tweetskb_polarized_users.iterrows():
        tsv_writer.writerow(row.values.tolist())
```

## The number of polarized and neutral users

```python
path = '../data/tweetskb_political_polarized_users_with_atleast_one_polarized_domain.tsv'

df_tweetskb_political_polarized_and_neutralized = pd.read_csv(path, sep='\t', error_bad_lines=False, names=['Username', 'Number_of_liberals', 'Number_of_conservatives', 'Sum_polarised_domains', 'political_polarisation'])

print('Shape :', df_tweetskb_political_polarized_and_neutralized.shape)
df_tweetskb_political_polarized_and_neutralized.head()
```

```python
df_tweetskb_political_polarized = df_tweetskb_political_polarized_and_neutralized[df_tweetskb_political_polarized_and_neutralized['political_polarisation'] != 0]

print('Shape: ', df_tweetskb_political_polarized.shape)
df_tweetskb_political_polarized.head()
```

```python
df_tweetskb_political_neutralized = df_tweetskb_political_polarized_and_neutralized[df_tweetskb_political_polarized_and_neutralized['political_polarisation'] == 0]

print('Shape of neutralized dataframe: ', df_tweetskb_political_neutralized.shape)
df_tweetskb_political_neutralized.head()
```

```python
path = '../data/tweetskb_science_polarized_users_with_atleast_one_polarized_domain.tsv'

df_tweetskb_science_polarized_and_neutralized = pd.read_csv(path, sep='\t', error_bad_lines=False, names=['Username', 'Number_of_antiscience_domains', 'Number_of_proscience_domains', 'Sum_polarised_domains', 'science_polarisation'])

print('Shape :', df_tweetskb_science_polarized_and_neutralized.shape)
df_tweetskb_science_polarized_and_neutralized.head()
```

```python
df_tweetskb_science_polarized = df_tweetskb_science_polarized_and_neutralized[df_tweetskb_science_polarized_and_neutralized['science_polarisation'] != 0]

print('Shape: ', df_tweetskb_science_polarized.shape)
df_tweetskb_science_polarized.head()
```

```python
df_tweetskb_science_neutralized = df_tweetskb_science_polarized_and_neutralized[df_tweetskb_science_polarized_and_neutralized['science_polarisation'] == 0]

print('Shape of neutralized dataframe: ', df_tweetskb_science_neutralized.shape)
df_tweetskb_science_neutralized.head()
```

```python
path = '../data/tweetskb_moderacy_polarized_users_with_atleast_one_polarized_domain.tsv'

df_tweetskb_moderacy_polarized_and_neutralized = pd.read_csv(path, sep='\t', error_bad_lines=False, names=['Username', 'Number_of_moderates', 'Number_of_hardlines', 'Sum_polarised_domains', 'moderacy_polarisation'])

print('Shape :', df_tweetskb_moderacy_polarized_and_neutralized.shape)
df_tweetskb_moderacy_polarized_and_neutralized.head()
```

```python
df_tweetskb_moderacy_polarized = df_tweetskb_moderacy_polarized_and_neutralized[df_tweetskb_moderacy_polarized_and_neutralized['moderacy_polarisation'] != 0]

print('Shape: ', df_tweetskb_moderacy_polarized.shape)
df_tweetskb_moderacy_polarized.head()
```

```python
df_tweetskb_moderacy_neutralized = df_tweetskb_moderacy_polarized_and_neutralized[df_tweetskb_moderacy_polarized_and_neutralized['moderacy_polarisation'] == 0]

print('Shape of neutralized dataframe: ', df_tweetskb_moderacy_neutralized.shape)
df_tweetskb_moderacy_neutralized.head()
```

```python
df_tweetskb_political_neutralized = df_tweetskb_political_neutralized[['Username', 'political_polarisation']]
df_tweetskb_science_neutralized = df_tweetskb_science_neutralized[['Username', 'science_polarisation']]
df_tweetskb_moderacy_neutralized = df_tweetskb_moderacy_neutralized[['Username', 'moderacy_polarisation']]

df_tweetskb_user_neutralized = pd.merge(df_tweetskb_political_neutralized, df_tweetskb_science_neutralized, on='Username', how='outer')
df_tweetskb_user_neutralized = pd.merge(df_tweetskb_user_neutralized, df_tweetskb_moderacy_neutralized, on='Username', how='outer')


print('Form :', df_tweetskb_user_neutralized.shape)
df_tweetskb_user_neutralized.head()
```

```python
df_tweetskb_political_polarized_and_neutralized = df_tweetskb_political_polarized_and_neutralized[['Username', 'political_polarisation']]
df_tweetskb_science_polarized_and_neutralized = df_tweetskb_science_polarized_and_neutralized[['Username', 'science_polarisation']]
df_tweetskb_moderacy_polarized_and_neutralized = df_tweetskb_moderacy_polarized_and_neutralized[['Username', 'moderacy_polarisation']]

df_tweetskb_user_polarized_and_neutralized = pd.merge(df_tweetskb_political_polarized_and_neutralized, df_tweetskb_science_polarized_and_neutralized, on='Username', how='outer')
df_tweetskb_user_polarized_and_neutralized = pd.merge(df_tweetskb_user_polarized_and_neutralized, df_tweetskb_moderacy_polarized_and_neutralized, on='Username', how='outer')


print('Form :', df_tweetskb_user_polarized_and_neutralized.shape)
df_tweetskb_user_polarized_and_neutralized.head()
```
