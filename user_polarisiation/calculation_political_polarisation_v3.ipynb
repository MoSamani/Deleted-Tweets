```python
def get_number_of_liberal_domains(domains):
    
    domains_list = domains.split()
    score = 0
    l = []
    for i in range(0, len(domains_list)):
        
        #d = tldextract.extract(domains_list[i])
        #d = d.domain
        d = domains_list[i]
        
        if d in list_liberal_domains:
            score -=  1
            l.append(d)
    return l
```

```python
df_tweetscov19['Domains'] = df_tweetscov19['Domains'].astype(str)

df_tweetscov19['Liberal_domains'] = df_tweetscov19.apply(lambda row: get_number_of_liberal_domains(row['Domains']) if (row['Domains'] != '0') else 0, axis=1)
```

```python
df_tweetscov19[df_tweetscov19.Number_of_liberals >1].head()
```

```python
def get_number_of_conservative_domains(domains):
    domains_list = domains.split()
    score = 0
    l = []
    for i in range(0, len(domains_list)):
        
        #d = tldextract.extract(domains_list[i])
        #d = d.domain
        d =  domains_list[i]
        
        if d in list_conservative_domains:
            score +=  1
            l.append(d)
    return l
```

```python
df_tweetscov19['Conservative_domains'] = df_tweetscov19.apply(lambda row: get_number_of_conservative_domains(row['Domains']) if (row['Domains'] != '0') else 0, axis=1)
```

```python
df_tweetscov19[df_tweetscov19.Number_of_conservatives > 1].head()
```

```python
path_tweetscov19 = '../data/tweetscov19_domainsLR_political_polarized_useres.tsv'

df_tweetscov19 = pd.read_csv(path_tweetscov19, sep='\t', error_bad_lines=False, names=['Tweet_Id', 'Username', 'Is_deleted', 'Domains', 'Liberal_domains', 'Number_of_liberals', 'Conservative_domains', 'Number_of_conservatives'])

print('Shape :', df_tweetscov19.shape)
df_tweetscov19.head()
```

```python
df_tweetscov19_del = df_tweetscov19[df_tweetscov19['Is_deleted'] == True]
df_tweetscov19_undel = df_tweetscov19[df_tweetscov19['Is_deleted'] == False]
```

```python
import ast

def get_unique_domains_with_their_number_of_repeatation(df):
    dict_domains = {}
    i  = 0

    for idx, row in df.iterrows():
        list_domains = []
        list_domains = (row.values.tolist()[0])
        
        list_domains = ast.literal_eval(list_domains)

        if (list_domains != 0):
            if (len(list_domains) > 0):
                #print(list_domains)
                for i in range(0, len(list_domains)):
                    domain = list_domains[i]
                    if domain not in dict_domains:
                        dict_domains[domain] = 1
                    else :
                        dict_domains[domain] = dict_domains[domain] +1
    return dict_domains
```

```python
df_tweetsCov19_domains_l = df_tweetscov19[['Liberal_domains']]
dict_tweetscov19_domains_l = get_unique_domains_with_their_number_of_repeatation(df_tweetsCov19_domains_l)
```

```python
df_dict_tweetsCov19_liberal_domains = pd.DataFrame(dict_tweetscov19_domains_l.items(), columns=['liberal_domains','Number'])
df_dict_tweetsCov19_liberal_domains.head(1)
```

```python
Number_of_liberal_domains = df_tweetscov19['Number_of_liberals'].sum()
print('Number of used liberal Domains in TweetsCov19(unique) :', df_dict_tweetsCov19_liberal_domains.liberal_domains.size) #-1 is label 0 for no Domains
print('Number of used liberal Domains in TweetsCov19(sum) :', Number_of_liberal_domains)
```

```python
df_dict_tweetsCov19_liberal_domains['percent'] =  df_dict_tweetsCov19_liberal_domains['Number'].apply(lambda row : ((row /2508542) * 100) )

df_dict_tweetsCov19_liberal_domains['Number'] = df_dict_tweetsCov19_liberal_domains['Number'].astype(int) 
df_dict_tweetsCov19_liberal_domains.sort_values(by = ['Number'], ascending = False, inplace = True)
df_dict_tweetsCov19_liberal_domains.head(10)
```

```python
df_tweetsCov19_domains_c = df_tweetscov19[['Conservative_domains']]
dict_tweetscov19_domains_c = get_unique_domains_with_their_number_of_repeatation(df_tweetsCov19_domains_c)
```

```python
df_dict_tweetsCov19_conservative_domains = pd.DataFrame(dict_tweetscov19_domains_c.items(), columns=['Conservative_domains','Number'])
```

```python
Number_of_conservatives = df_tweetscov19['Number_of_conservatives'].sum()
print('Number of used conservative Domains in TweetsCov19(unique) :', df_dict_tweetsCov19_conservative_domains.Conservative_domains.size) #-1 is label 0 for no Domains
print('Number of used conservative Domains in TweetsCov19(sum) :', Number_of_conservatives)
```

```python
df_dict_tweetsCov19_conservative_domains['percent'] =  df_dict_tweetsCov19_conservative_domains['Number'].apply(lambda row : ((row /2508542) * 100) )

df_dict_tweetsCov19_conservative_domains['Number'] = df_dict_tweetsCov19_conservative_domains['Number'].astype(int) 
df_dict_tweetsCov19_conservative_domains.sort_values(by = ['Number'], ascending = False, inplace = True)
df_dict_tweetsCov19_conservative_domains.head(10)
```

```python
path_tweetscov19 = '../data/tweetscov19_domainsLR_political_polarized_useres.tsv'

df_tweetscov19 = pd.read_csv(path_tweetscov19, sep='\t', error_bad_lines=False, names=['Tweet_Id', 'Username', 'Is_deleted', 'Domains', 'Liberal_domains', 'Number_of_liberals', 'Conservative_domains', 'Number_of_conservatives'])

print('Shape :', df_tweetscov19.shape)
df_tweetscov19.head()
```

```python
cond1 = df_tweetscov19.Number_of_liberals != 0
cond2 = df_tweetscov19.Number_of_conservatives != 0
df_tweetscov19 = df_tweetscov19[(cond1 | cond2)]
```

```python
df_tweetscov19.Number_of_liberals = df_tweetscov19.Number_of_liberals *-1
```

```python
df_tweetscov19['tweet_political_polarisation'] = df_tweetscov19.apply(lambda row: ( ((row['Number_of_liberals'] + row['Number_of_conservatives']) / (row['Number_of_conservatives'] - row['Number_of_liberals'])) if (row['Number_of_conservatives'] - row['Number_of_liberals']) != 0 else 0), axis=1)
```

```python
df_tweetscov19.describe()
```

```python
df_tweetscov19[df_tweetscov19.tweet_political_polarisation < 0 ].head()
```
