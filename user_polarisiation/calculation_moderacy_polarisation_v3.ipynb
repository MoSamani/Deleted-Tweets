```python
def get_number_of_moderate_domains(domains):
    domains_list = domains.split()
    score = 0
    l = []
    for i in range(0, len(domains_list)):
        
        #d = tldextract.extract(domains_list[i])
        #d = d.domain
        
        d = domains_list[i]
        
        if d in list_moderat_domains:
            score -=  1
            l.append(d)
        
    return l
```

```python
df_tweetscov19['Domains'] = df_tweetscov19['Domains'].astype(str)

df_tweetscov19['Moderate_domains'] = df_tweetscov19.apply(lambda row: get_number_of_moderate_domains(row['Domains']) if (row['Domains'] != '0') else 0, axis=1)
```

```python
df_tweetscov19.head()
```

```python
df_tweetscov19['Number_of_moderate_domains'] = df_tweetscov19.apply(lambda row: len(row['Moderate_domains']) if (row['Moderate_domains'] != 0) else 0, axis=1)
```

```python
df_tweetscov19[df_tweetscov19.Number_of_moderate_domains > 1].head()
```

```python
def get_number_of_hardline_domains(domains):
    domains_list = domains.split()
    score = 0
    l = []
    for i in range(0, len(domains_list)):
        
        #d = tldextract.extract(domains_list[i])
        #d = d.domain
       
        d = domains_list[i]
        
        if d in list_hardline_domains:
            score +=  1
            l.append(d)
        
    return l
```

```python
df_tweetscov19['Hardline_domains'] = df_tweetscov19.apply(lambda row: get_number_of_hardline_domains(row['Domains']) if (row['Domains'] != '0') else 0, axis=1)
```

```python
df_tweetscov19.head()
```

```python
df_tweetscov19[df_tweetscov19.Number_of_hardline_domains > 1].head()
```

```python
path_tweetscov19 = '../data/tweetscov19_domainsLR_moderacy_polarized_useres.tsv'

df_tweetscov19 = pd.read_csv(path_tweetscov19, sep='\t', error_bad_lines=False, names=['Tweet_Id', 'Username', 'Is_deleted', 'Domains', 'Moderate_domains', 'Number_of_moderate_domains', 'Hardline_domains', 'Number_of_hardline_domains'])

print('Shape :', df_tweetscov19.shape)
df_tweetscov19.head()
```

```python
df_tweetscov19_del = df_tweetscov19[df_tweetscov19['Is_deleted'] == True]
df_tweetscov19_undel = df_tweetscov19[df_tweetscov19['Is_deleted'] == False]
```

```python
import ast

def get_unique_domains_with_their_number_of_repeatation(df):
    dict_domains = {}
    i  = 0

    for idx, row in df.iterrows():
        list_domains = []
        list_domains = (row.values.tolist()[0])
        
        list_domains = ast.literal_eval(list_domains)

        if (list_domains != 0):
            if (len(list_domains) > 0):
                #print(list_domains)
                for i in range(0, len(list_domains)):
                    domain = list_domains[i]
                    if domain not in dict_domains:
                        dict_domains[domain] = 1
                    else :
                        dict_domains[domain] = dict_domains[domain] +1
    return dict_domains
```

```python
df_tweetsCov19_domains_m = df_tweetscov19[['Moderate_domains']]
dict_tweetscov19_domains_m = get_unique_domains_with_their_number_of_repeatation(df_tweetsCov19_domains_m)
```

```python
df_dict_tweetsCov19_moderate_domains = pd.DataFrame(dict_tweetscov19_domains_m.items(), columns=['Moderate_domains','Number'])
df_dict_tweetsCov19_moderate_domains.head(1)
```

```python
Number_of_moderate_domains = df_tweetscov19['Number_of_moderate_domains'].sum()
print('Number of used moderate Domains in TweetsCov19(unique) :', df_dict_tweetsCov19_moderate_domains.Moderate_domains.size)
print('Number of used moderate Domains in TweetsCov19(sum) :', Number_of_moderate_domains)
```

```python
df_dict_tweetsCov19_moderate_domains['percent'] =  df_dict_tweetsCov19_moderate_domains['Number'].apply(lambda row : ((row /2508542) * 100) )

df_dict_tweetsCov19_moderate_domains['Number'] = df_dict_tweetsCov19_moderate_domains['Number'].astype(int) 
df_dict_tweetsCov19_moderate_domains.sort_values(by = ['Number'], ascending = False, inplace = True)
df_dict_tweetsCov19_moderate_domains.head(10)
```

```python
df_tweetsCov19_domains_h = df_tweetscov19[['Hardline_domains']]
dict_tweetscov19_domains_h = get_unique_domains_with_their_number_of_repeatation(df_tweetsCov19_domains_h)
```

```python
df_dict_tweetsCov19_hardliner_domains = pd.DataFrame(dict_tweetscov19_domains_h.items(), columns=['Hardliner_domains','Number'])
```

```python
Number_of_hardliner = df_tweetscov19['Number_of_hardline_domains'].sum()
print('Number of used hardliner Domains in TweetsCov19(unique) :', df_dict_tweetsCov19_hardliner_domains.Hardliner_domains.size) 
print('Number of used hardliner Domains in TweetsCov19(sum) :', Number_of_hardliner)
```

```python
df_dict_tweetsCov19_hardliner_domains['percent'] =  df_dict_tweetsCov19_hardliner_domains['Number'].apply(lambda row : ((row /2508542) * 100) )

df_dict_tweetsCov19_hardliner_domains['Number'] = df_dict_tweetsCov19_hardliner_domains['Number'].astype(int) 
df_dict_tweetsCov19_hardliner_domains.sort_values(by = ['Number'], ascending = False, inplace = True)
df_dict_tweetsCov19_hardliner_domains.head(10)
```

```python
path_tweetscov19 = '../data/tweetscov19_domainsLR_moderacy_polarized_useres.tsv'

df_tweetscov19 = pd.read_csv(path_tweetscov19, sep='\t', error_bad_lines=False, names=['Tweet_Id', 'Username', 'Is_deleted', 'Domains', 'Moderate_domains', 'Number_of_moderate_domains', 'Hardline_domains', 'Number_of_hardline_domains'])

print('Shape :', df_tweetscov19.shape)
df_tweetscov19.head()
```

```python
cond1 = df_tweetscov19.Number_of_moderate_domains != 0
cond2 = df_tweetscov19.Number_of_hardline_domains != 0
df_tweetscov19 = df_tweetscov19[(cond1 | cond2)]
```

```python
df_tweetscov19.Number_of_moderate_domains = df_tweetscov19.Number_of_moderate_domains *-1
```

```python
df_tweetscov19['tweet_moderacy_polarisation'] = df_tweetscov19.apply(lambda row: ( (row['Number_of_moderate_domains'] + row['Number_of_hardline_domains']) / ( row['Number_of_hardline_domains'] - row['Number_of_moderate_domains']) if ( row['Number_of_hardline_domains'] - row['Number_of_moderate_domains']) != 0 else 0 ), axis=1)
```
