```python
path = '../data/tweetscov19_id_isdeleted_domains.tsv'

df_tweetscov19 = pd.read_csv(path, sep='\t', error_bad_lines=False, names=['Tweet_Id', 'Username' ,'Is_deleted', 'Domains'])

print('Shape :', df_tweetscov19.shape)
df_tweetscov19.head()
```

```python
def get_number_of_antiscience_domains(domains):
    domains_list = domains.split()
    score = 0
    l = []
    for i in range(0, len(domains_list)):
        
        #d = tldextract.extract(domains_list[i])
        #d = d.domain          
        d = domains_list[i]
        
        if d in list_antiscience_domains:
            score -=  1
            l.append(d)
        
    return l
```

```python
df_tweetscov19['Domains'] = df_tweetscov19['Domains'].astype(str)

df_tweetscov19['Antisciens_domains'] = df_tweetscov19.apply(lambda row: get_number_of_antiscience_domains(row['Domains']) if (row['Domains'] != '0') else 0, axis=1)
```

```python
df_tweetscov19.head()
```

```python
df_tweetscov19['number_of_antisciens_domains'] = df_tweetscov19.apply(lambda row: len(row['Antisciens_domains']) if (row['Antisciens_domains'] != 0) else 0, axis=1)
```

```python
df_tweetscov19[df_tweetscov19.number_of_antisciens_domains > 1].head()
```

```python
def get_number_of_proscience_domains(domains):
    domains_list = domains.split()
    score = 0
    l = []
    for i in range(0, len(domains_list)):
        
        #d = tldextract.extract(domains_list[i])
        #d = d.domain
       
            
        d = domains_list[i]
        if d in list_proscience_domains:
            score +=  1
            l.append(d)
        
    return l
```

```python
df_tweetscov19['Prosciens_domains'] = df_tweetscov19.apply(lambda row: get_number_of_proscience_domains(row['Domains']) if (row['Domains'] != '0') else 0, axis=1)
```

```python
df_tweetscov19.head()
```

```python
df_tweetscov19['Number_of_prosciens_domains'] = df_tweetscov19.apply(lambda row: len(row['Prosciens_domains']) if (row['Prosciens_domains'] != 0) else 0, axis=1)
```

```python
df_tweetscov19[df_tweetscov19.number_of_prosciens_domains > 1].head()
```

```python
path_tweetscov19 = '../data/tweetscov19_domainsLR_science_polarized_useres.tsv'

df_tweetscov19 = pd.read_csv(path_tweetscov19, sep='\t', error_bad_lines=False, names=['Tweet_Id', 'Username', 'Is_deleted', 'Domains', 'Antisciens_domains', 'number_of_antisciens_domains', 'Prosciens_domains', 'number_of_prosciens_domains'])

print('Shape :', df_tweetscov19.shape)
df_tweetscov19.head()
```

```python
df_tweetscov19_del = df_tweetscov19[df_tweetscov19['Is_deleted'] == True]
df_tweetscov19_undel = df_tweetscov19[df_tweetscov19['Is_deleted'] == False]
```

```python
import ast

def get_unique_domains_with_their_number_of_repeatation(df):
    dict_domains = {}
    i  = 0

    for idx, row in df.iterrows():
        list_domains = []
        list_domains = (row.values.tolist()[0])
        
        list_domains = ast.literal_eval(list_domains)

        if (list_domains != 0):
            if (len(list_domains) > 0):
                #print(list_domains)
                for i in range(0, len(list_domains)):
                    domain = list_domains[i]
                    if domain not in dict_domains:
                        dict_domains[domain] = 1
                    else :
                        dict_domains[domain] = dict_domains[domain] +1
    return dict_domains
```

```python
df_tweetsCov19_domains_asc = df_tweetscov19[['Antisciens_domains']]
dict_tweetscov19_domains_asc = get_unique_domains_with_their_number_of_repeatation(df_tweetsCov19_domains_asc)
```

```python
df_dict_tweetsCov19_antiscience_domains = pd.DataFrame(dict_tweetscov19_domains_asc.items(), columns=['Antiscience_domains','Number'])
df_dict_tweetsCov19_antiscience_domains.head(1)
```

```python
Number_of_antiscience_domains = df_tweetscov19['number_of_antisciens_domains'].sum()
print('Number of used antiscience Domains in TweetsCov19(unique) :', df_dict_tweetsCov19_antiscience_domains.Antiscience_domains.size)
print('Number of used antiscience Domains in TweetsCov19(sum) :', Number_of_antiscience_domains)
```

```python
df_dict_tweetsCov19_antiscience_domains['percent'] =  df_dict_tweetsCov19_antiscience_domains['Number'].apply(lambda row : ((row /2508542) * 100) )

df_dict_tweetsCov19_antiscience_domains['Number'] = df_dict_tweetsCov19_antiscience_domains['Number'].astype(int) 
df_dict_tweetsCov19_antiscience_domains.sort_values(by = ['Number'], ascending = False, inplace = True)
df_dict_tweetsCov19_antiscience_domains.head(10)
```

```python
df_tweetsCov19_domains_psc = df_tweetscov19[['Prosciens_domains']]
dict_tweetscov19_domains_psc = get_unique_domains_with_their_number_of_repeatation(df_tweetsCov19_domains_psc)
```

```python
df_dict_tweetsCov19_proscience_domains = pd.DataFrame(dict_tweetscov19_domains_psc.items(), columns=['Proscience_domains','Number'])
```

```python
Number_of_proscience = df_tweetscov19['number_of_prosciens_domains'].sum()
print('Number of used proscience Domains in TweetsCov19(unique) :', df_dict_tweetsCov19_proscience_domains.Proscience_domains.size) 
print('Number of used proscience Domains in TweetsCov19(sum) :', Number_of_proscience)
```

```python
df_dict_tweetsCov19_proscience_domains['percent'] =  df_dict_tweetsCov19_proscience_domains['Number'].apply(lambda row : ((row /2508542) * 100) )

df_dict_tweetsCov19_proscience_domains['Number'] = df_dict_tweetsCov19_proscience_domains['Number'].astype(int) 
df_dict_tweetsCov19_proscience_domains.sort_values(by = ['Number'], ascending = False, inplace = True)
df_dict_tweetsCov19_proscience_domains.head(10)
```

```python
path_tweetscov19 = '../data/tweetscov19_domainsLR_science_polarized_useres.tsv'

df_tweetscov19 = pd.read_csv(path_tweetscov19, sep='\t', error_bad_lines=False, names=['Tweet_Id', 'Username', 'Is_deleted', 'Domains', 'Antisciens_domains', 'number_of_antisciens_domains', 'Prosciens_domains', 'number_of_prosciens_domains'])

print('Shape :', df_tweetscov19.shape)
df_tweetscov19.head()
```
